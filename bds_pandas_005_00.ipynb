 # **<font color="#07a8ed">Introducci贸n a las agrupaciones</font>**

````python
adult_census.head(3)  ## encabezado de 3 filas del dataset
````

<p align="justify">  Vamos a agrupar el <code>DataFrame</code> por edad...  </p>

````python
edades = adult_census["age"].groupby(adult_census["age"])  ### genero la variables edades que surge de la columna edad agrupada
edades
````

<p align="justify">  La variable edades es ahora un objeto especial. En realidad, todav铆a no ha calculado nada. La idea es que este objeto tenga toda la informaci贸n necesaria para luego aplicar alguna operaci贸n..., como por ejemplo, contar la cantidad de edades...  </p>

````python
type(edades)
````

````python
edades.count()  ### contar el agurpamiento por edades
````

<p align="justify">  La indexaci贸n de un objeto <code>GroupBy</code> creado a partir de un <code>DataFrame</code> con un nombre de columna tiene el efecto de crear un subconjunto para la agregaci贸n..., que en este caso es el c谩lculo de la media de las horas por semana, en virtud de las edades...</p>

````python
adult_census[['age', "hours-per-week"]]  ### imprimo las dos columnas originales del dataFrame
````

````python
px.histogram(adult_census[['age', "hours-per-week"]], x="hours-per-week", nbins=20)  ## histograma de datos originales
````

 # **<font color="#07a8ed">Aplicaci贸n de funciones m煤ltiples y en columnas</font>**

<p align="justify">  Vamos a ver un conjunto de datos con las siguientes columnas:
</p>

- total_bill es el total facturado expresado en dolares
- tip es la propina en dolares
- sex es el genero de quien paga
- smoker es si hab铆an fumadores
- day es el dia de la semana
- time es para saber si es cena o almuerzo
- size es la cantidad de personas en el evento


````python
tips = pd.read_csv("https://raw.githubusercontent.com/cristiandarioortegayubro/BDS/main/datasets/tips.csv")
tips
````

<p align="justify">  Vamos a agregar una columna para calcular que porcentaje tiene la propina sobre el pago total:


````python
tips["tip_pct"] = (tips["tip"] / tips["total_bill"]).round(3)  ###   creo una columna nueva tip_pct su contenido es % total de propina / total de cuenta
````

````python
tips
````

````python
tips.info()
````

<p align="justify">  Ahora vamos a agrupar las propinas por <code>day</code> y por <code>smoker</code>:


````python
grouped = tips.groupby(["day", "smoker"])
````

<p align="justify">  Vamos a agregar una <code>Serie</code> a un <code>DataFrame</code> usando <code>aggregate()</code> con la funci贸n deseada o llamando a un m茅todo como <code>mean()</code> o <code>std()</code>:

````python
grouped_pct = grouped["tip_pct"]  ## creo una serie con la nueva columna
````

````python
grouped_pct.agg("mean").round(2)  ## agregue columna con la media de la nueva columna me quedan 3 columnas
````

<p align="justify">  Si pasamos una lista de m茅todos <code>mean()</code> o <code>std()</code>, obtenemos un <code>DataFrame</code> con esos nombres de columnas:

````python
grouped_pct.agg(["mean", "std"]).round(2) ## agregue a la media la desviacion estandar
````

<p align="justify"> N贸tese que el <code>DataFrame</code> es de 铆ndice multiple, con dos niveles de jerarquias.

<p align="justify">  Con un <code>DataFrame</code> hay m谩s opciones, ya que se puede especificar una lista con funciones para aplicar a todas las columnas, o tambien se puede definir diferentes funciones para cada columna:

````python
funciones = ["count", "mean", "max"]
````

````python
resultados = grouped[["tip_pct", "total_bill"]].agg(funciones).round(3)  ## agregue a la media la desviacion estandar de estas columnas numericas
````

````python
resultados
````

<p align="justify"> Este es un <code>DataFrame</code> con 铆ndice y columnas multiples, con dos niveles de jerarquias tanto en los 铆ndices como en las columnas. Si solo quiero visualizar la columna <code>tip_pct</code>, entonces debemos usar el nombre de la columna como argumento, por ejemplo:

````python
resultados["tip_pct"]  # detallando columnas
````

````python
resultados["total_bill"]
````

<p align="justify">  Ahora queremos aplicar diferentes funciones para cada columna en un <code>DataFrame</code>, lo que lo logramos generando un diccionario. Para ello, en las claves del diccionario se deben colocar los nombres de las columnas que se desea aplicar alguna funci贸n:

````python
tips  ## dataset original
````

````python
grouped.agg({"tip" : np.max, "size" : "sum"}) ## en una columna aplico el maximo y en la otra la suma de los valores
````

````python
grouped.agg({"tip_pct" : ["min", "max", "mean", "std"],  ## en una columna aplico min max media y desviacion estanadar en la otra la suma en forma de diccionario
             "size" : "sum"}).round(3)
````

 ## **<font color="#07a8ed">Datos agregados sin 铆ndices de fila</font>**

<p align="justify">  Podemos deshabilitar el 铆ndice con el par谩metro <code> as_index =False</code>, entonces nuestro <code>DataFrame</code> no tendr铆a un 铆ndice m煤ltiple, sino que posee el siguiente aspecto:

````python
grouped = tips.groupby(["day", "smoker"], as_index=False)  ## agrupop por las variables day y smoker y deshabilito el indice as_index = False
````

````python
grouped.mean(numeric_only=True).round(3) ## agrupo por numericas y su media redondeando a 3 decimales
````

 ## **<font color="#07a8ed">Usando apply( ) en groupby( )</font>**

<p align="justify">  Primero vamos a generar una funci贸n que calcule el top $5$ en propinas, es decir las $5$ propinas mas 谩ltas:

````python
def top(df, n=5, column="tip_pct"):  ## funcion de cobranza de las 5 propinas mas altas
  return df.sort_values(column, ascending=False)[:n]  ### return de valores sort port las columnas en forma descendente
````

<p align="justify">  Ahora aplicamos la funci贸n a nuestro <code>DataFrame</code>:

````python
tips
````

````python
top(tips, n=6)  ## aplico la funcion al dataset tips
````

<p align="justify">  Ahora, si agrupamos por <code>smoker</code>, y usamos <code>apply</code> con la funci贸n que acabamos de crear, entonces obtenemos lo siguiente:

````python
tips
````

````python
tips.groupby("smoker").apply(top)  ### agrupamos por smoker aplicando al funcion top
````

<p align="justify">  Y tambien, se pueden colocar otros argumentos en la funci贸n:

````python
tips.groupby(["smoker", "day"]).apply(top, n=1, column="total_bill") ### se agruopa por smoker y day y se aplica la funcion top con la propina alta en funcion de total_bill
````

<p align="justify">  Otro ejemplo:

````python
result = tips.groupby("smoker")["tip_pct"].describe().round(3)  ### creo variable resultado y le asigno la agrupacion por smoker y tip_pct calculo las metricas estadisticas
````
