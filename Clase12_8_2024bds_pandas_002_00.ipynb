<p align="justify">
<mark> La alineaci칩n de datos internos introduce valores faltantes en las ubicaciones de las etiquetas que no se superponen</mark>, es decir, aquellas series que tienen m치s elemenos, o elementos que no est치n en otras series.</p>

<p align="justify"> 游 En el caso de los <code>DataFrame</code>, la alineaci칩n se realiza tanto en filas como en columnas:
</p>

```python
ventas1 = pd.Series([1200, 4250, 3090, 3400], index=[2019,2020,2022,2023])
costo1 = pd.Series([800, 1900, 1200, 1450], index=[2019,2020,2022,2023])
bruto1 = ventas1 - costo1
```

```python
ventas2 = pd.Series([4250, 2050, 3400], index=[2020,2021,2023])
costo2 = pd.Series([1900, 1000, 1450], index=[2020,2021,2023])
bruto2 = ventas2 - costo2
```

```python
sucursal1 = {"ventas":ventas1,"costo":costo1,"bruto":bruto1}
sucursal2 = {"ventas":ventas2,"costo":costo2,"bruto":bruto2}
```

```python
sucursal1 = pd.DataFrame(sucursal1).T
sucursal2 = pd.DataFrame(sucursal2).T
```

```python
sucursal1
```

```python
sucursal2
```

```python
sucursal1 + sucursal2
```

 # **<font color="#07a8ed">Correlaci칩n y covarianza</font>**

<p align="justify">
Algunas estad칤sticas de resumen, como la correlaci칩n y la covarianza, se calculan a partir de argumentos. </p>

```python
resultado = sucursal1.T
resultado
```

<p align="justify">
Algunas estad칤sticas de resumen, como la correlaci칩n y la covarianza, se calculan a partir de argumentos. </p>

<p align="justify">
游 Podemos calcular la correlaci칩n con el m칠todo <code>corr()</code> entre las ventas y los costos.
</p>

```python
resultado["ventas"].corr(resultado["costo"]).round(2)
```

```python
resultado.ventas.corr(resultado.costo).round(2)
```

<p align="justify">
La correlaci칩n es una medida estad칤stica que indica la fuerza y la direcci칩n de la relaci칩n entre dos variables. Se mide con el coeficiente de correlaci칩n, que puede variar entre -1 y 1:
<br><br>

- **1** indica una correlaci칩n positiva perfecta: cuando una variable aumenta, la otra tambi칠n lo hace en proporci칩n directa.
- **-1** indica una correlaci칩n negativa perfecta: cuando una variable aumenta, la otra disminuye en proporci칩n directa.
- **0** indica que no hay correlaci칩n lineal entre las variables.

<p align="justify">
<br>
En el script <code>resultado["ventas"].corr(resultado["costo"])</code>, se est치 calculando el coeficiente de correlaci칩n entre dos columnas de un DataFrame de pandas: <code>ventas</code> y <code>costo</code>.
<br><br>

- **`resultado["ventas"]`**: Es una serie que representa las ventas en un conjunto de datos.
- **`resultado["costo"]`**: Es una serie que representa los costos en el mismo conjunto de datos.

<p align="justify">
<br>
La funci칩n <code>corr()</code> calcula el coeficiente de correlaci칩n entre estas dos series. El valor resultante te dir치 c칩mo est치n relacionadas las ventas y los costos:
<br><br>

- Si el coeficiente es cercano a 1, hay una fuerte relaci칩n positiva: a mayor costo, mayores ventas.
- Si el coeficiente es cercano a -1, hay una fuerte relaci칩n negativa: a mayor costo, menores ventas.
- Si el coeficiente es cercano a 0, no hay una relaci칩n lineal significativa entre ventas y costos.



<p align="justify">
游 Tambi칠n podemos calcular la covarianza con el m칠todo <code>cov()</code> entre las ventas y los costos.
</p>

```python
resultado["ventas"].cov(resultado["costo"]).round(2)  # la covarianza es la direccion de la relacion de 2 variables
```

```python
resultado.ventas.cov(resultado.costo).round(2)
```
