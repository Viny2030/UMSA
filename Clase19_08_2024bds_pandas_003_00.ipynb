 ## **<font color="#07a8ed">Lectura de archivos en partes</font>**

<p align="justify">
Al procesar archivos muy grandes o descubrir el conjunto correcto de argumentos para procesar correctamente un archivo grande, <code>Pandas</code> permite leer solo una peque帽a parte de un archivo o iterar a trav茅s de partes m谩s peque帽as del archivo.</p>

```python
datos="https://raw.githubusercontent.com/cristiandarioortegayubro/BDS/main/datasets/beneficios-otorgados-colaborativa-202210.xlsx-Dataset.csv"
```

```python
subsidios = pd.read_csv(datos)
subsidios.head()
```

<p align="justify">  Los puntos suspensivos $...$ indican que se han omitido las filas en el medio del <code>DataFrame</code>. Se puede leer solo una peque帽a cantidad de filas (evitando leer todo el archivo), con el par谩metro <code>nrows</code> indicando en 茅l, la cantidad de filas o 铆ndices que se quieren leer.
</p>

```python
subsidios ## los puntos suspensivos indican que hay filas que no se lograron leer
```

```python
pd.read_csv(datos, nrows=4)
```

<p align="justify">  Para leer un archivo en partes, se utiliza el parametro <code>chunksize</code> y se especifica el n煤mero de filas o 铆ndices.
</p>

```python
subsidios = pd.read_csv(datos, nrows=4)
subsidios
```

```python
subsidios1 = pd.read_csv(datos, nrows=4)
subsidios1
```

```python
subsidios1.shape
```

```python
partes = pd.read_csv(datos, chunksize=20) # dividir el dataset en partes iguales pueden fraccionarlo en partes
```

<p align="justify">
El script <code>partes = pd.read_csv(datos, chunksize=20)</code> utiliza la funci贸n <code>read_csv</code> de la biblioteca pandas para leer un archivo CSV en partes o "chunks" de 20 filas cada una, en lugar de cargar el archivo completo en memoria. Esto es 煤til para manejar archivos grandes de manera m谩s eficiente, permitiendo procesar los datos en fragmentos m谩s peque帽os y reducir el consumo de memoria.
<br><br>
La variable <code>partes</code> contendr谩 un iterador sobre estos fragmentos de datos.

```python
type(partes)
```

```python
beneficios = pd.Series([], dtype="int64")
```

```python
for i in partes:
    beneficios = beneficios.add(i["beneficio_tipo"].value_counts(), fill_value=0)
```

<p align="justify">
El script itera sobre cada fragmento de datos generado por un iterador de pandas y actualiza la serie <code>beneficios</code> sumando los conteos de valores 煤nicos de la columna <code>"beneficio_tipo"</code> de cada fragmento. El m茅todo <code>add</code> combina los conteos acumulados de cada fragmento, utilizando <code>fill_value=0</code> para manejar valores faltantes y asegurarse de que se sumen correctamente incluso si algunos fragmentos tienen categor铆as diferentes.
<br><br>
Esto permite consolidar los conteos de beneficio a lo largo de todo el conjunto de datos dividido en fragmentos.

```python
beneficios = beneficios.sort_values(ascending=False)
beneficios
```

<p align="justify">  En el caso de la funcion <code>read_csv()</code>, los par谩metros m谩s destacados, son los siguientes:
<ul>
<li><code>path()</code> indica la ubicaci贸n del sistema de archivos, URL u objeto similar a un archivo</li>
<li><code>sep()</code> caracteres o expresi贸n regular que se usar谩 para dividir campos en cada fila</li>
<li><code>header()</code> n煤mero de fila para nombres de columna, deber铆a ser <code>None</code> si no hay encabezado.</li>
<li><code>index_col()</code> n煤mero de columna o nombre de columna para usar como 铆ndice</li>
<li><code>names()</code> lista de nombres de columnas</li>
<li><code>skiprows()</code> n煤mero de filas al principio del archivo para ignorar o lista de n煤meros de fila (a partir de 0) para omitir</li>
<li><code>na_values()</code> secuencia de valores para reemplazar con NA</li>
<li><code>keep_default_na()</code> usar o no la lista de valores predeterminados NA (<code>True</code> por defecto)</li>
<li><code>parse_dates()</code> intenta analizar los datos en <code>datetime</code></li>
<li><code>converters()</code> aplicar铆a la funci贸n <code>f</code> a los valores indicados </li>
<li><code>dayfirst()</code> al analizar fechas ambiguas, se le aplica formato internacional</li>
<li><code>verbose()</code> muestra informaci贸n de an谩lisis diversa</li>
<li><code>encoding()</code> codificaci贸n de texto</li>
<li><code>squeeze()</code> Si los datos analizados contienen solo una columna, devuelva una <code>Serie</code></li>
<li><code>thousands()</code> separador de miles</li>
