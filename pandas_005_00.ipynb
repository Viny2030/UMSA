# **<font color="DarkBlue">Aspecto de los graficos en Pandas**

```python
pd.options.plotting.backend = "plotly"
```

<p align="justify">
El script <code>pd.options.plotting.backend = "plotly"</code> se utiliza para configurar el backend de visualizaci칩n de gr치ficos en la biblioteca <b>Pandas</b> a <b>Plotly</b>.
<br><br>
Pandas es una biblioteca de Python ampliamente utilizada para la manipulaci칩n y an치lisis de datos, y cuenta con capacidades de visualizaci칩n integradas que permiten crear gr치ficos de forma sencilla.
<br><br>
Por defecto, Pandas utiliza Matplotlib como su backend para graficar. Sin embargo, al establecer la opci칩n de backend a Plotly, los gr치ficos generados por Pandas se crear치n utilizando la biblioteca Plotly, que es conocida por su capacidad para crear visualizaciones interactivas y atractivas.
<br><br>
Al cambiar el backend a Plotly, se obtienen varias ventajas:
<br><br>
<ul>
    <li><b>Interactividad:</b> Los gr치ficos generados son interactivos, permitiendo al usuario acercarse, alejarse y explorar los datos de manera m치s efectiva.</li>
    <li><b>Visualizaciones atractivas:</b> Plotly proporciona un estilo visual m치s moderno y est칠tico en comparaci칩n con los gr치ficos est치ticos de Matplotlib.</li>
    <li><b>Facilidad de uso:</b> Se puede seguir utilizando la misma sintaxis de Pandas para graficar, lo que facilita la transici칩n a visualizaciones interactivas sin necesidad de aprender una nueva API.</li>
</ul>
<p align="justify">
<br>
En resumen, este script es una forma de aprovechar la potencia de Plotly para mejorar la experiencia de visualizaci칩n de datos en proyectos que utilizan Pandas para el an치lisis y manipulaci칩n de datos.
</p>


## **<font color="DarkBlue">Graficando con Pandas**

<p align="justify"> 游 Ejemplo:  
<br><br>
Generaremos un gr치fico de l칤neas a partir de un conjunto de datos de ventas mensuales.

```python
# Crear un DataFrame de ejemplo con datos de ventas mensuales
data = {
    'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre'],
    'Ventas': np.random.randint(1000, 5000, size=12)  # Ventas aleatorias entre 1000 y 5000
}

df = pd.DataFrame(data)
df
```

```python
# Crear un gr치fico de l칤neas de las ventas mensuales
df.plot("Mes","Ventas", template='gridon', title='Ventas anuales')
```

```python
# Crear un gr치fico de l칤neas de las ventas mensuales
df.plot(x='Mes', y='Ventas', title='Ventas Mensuales', markers=True, template='plotly_white')
```

## **<font color="DarkBlue">Tipos de gr치ficos**

```python
# Crear un DataFrame de ejemplo con datos aleatorios
np.random.seed(42)  # Para reproducibilidad
data = {
    'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre'],
    'Ventas': np.random.randint(1000, 5000, size=12),  # Ventas aleatorias entre 1000 y 5000
    'Costos': np.random.randint(500, 3000, size=12)    # Costos aleatorios entre 500 y 3000
}

df = pd.DataFrame(data)
df
```

### **<font color="DarkBlue">Gr치fico de Lineas**

```python
# 1. Gr치fico de L칤neas
fig_line = df.plot(x='Mes',
                   y=['Ventas'],
                   title='Ventas Mensuales',
                   markers=True,
                   template='plotly_white',
                   hover_name='Mes')

fig_line.show()
```

```python
df.plot(x='Mes',
        y=['Ventas',"Costos"],
        title='Ventas y Costos Mensuales',
        markers=True,
        template="plotly_white",
        labels={"variable": "Ventas y Costos"},
        hover_name="Mes")
```

### **<font color="White">Gr치fico de Barras**

```python
# 2. Gr치fico de Barras
df.plot("Mes",
        ["Ventas","Costos"],
        kind="bar",
        labels={"variable": "Ventas y costos"})
```

```python
# 2. Gr치fico de Barras
fig_bar = df.plot(x='Mes',
                  y='Ventas',
                  kind='bar',
                  title='Ventas Mensuales',
                  color='Mes',
                  template="plotly_white",
                  text="Ventas")
fig_bar.show()
```

```python
# 2. Gr치fico de Barras
fig_bar = df.plot(x='Mes',
                  y=['Ventas','Costos'],
                  kind='bar',
                  title='Ventas Mensuales',
                  color='Mes',
                  template="plotly_white")
                  #text="Ventas") #Commented out as it causes a syntax error
fig_bar.show()
```

```python
# 2. Gr치fico de Barras
df.plot("Ventas","Mes", kind="bar")
```

```python
# 2. Gr치fico de Barras
fig_bar = df.plot(y='Mes', x='Ventas', kind='bar', title='Ventas Mensuales', color='Mes')
fig_bar.update_layout(template="plotly_white")
fig_bar.show()
```

### **<font color="DarkBlue">Gr치fico Histograma**

```python
# 3. Histograma
df.plot("Ventas", kind="hist").update_layout(bargap=0.2)
```

```python
# 3. Histograma
fig_hist = df['Ventas'].plot.hist(title='Histograma de Ventas')
fig_hist.update_layout(template="plotly_white", bargap=0.2)
fig_hist.show()
```

```python
df['Ventas'].plot.hist(title="Histograma de Ventas", bins=10)
```

### **<font color="DarkBlue">Box Plot**

```python
# 4. Box Plot
df.plot(x="Ventas", kind="box")
```

```python
# 4. Box Plot
fig_box = df.plot.box(title='Box Plot de Ventas y Costos', x="Ventas")
fig_box.update_layout(template="plotly_white")
fig_box.show()
```

```python
# 4. Box Plot
fig_box = df.plot.box(title='Box Plot de Ventas y Costos', y="Ventas",template="plotly_white")

fig_box.show()
```

```python
# 4. Box Plot
fig_box = df.plot.box(title='Box Plot de Ventas y Costos', y="Ventas")
fig_box.update_layout(template="plotly_white")
fig_box.show()
```
