 ## **<font color="#07a8ed">Rellenado de datos faltantes</font>**

<p align="justify"> En lugar de filtrar datos faltantes, y tal vez descartar otros datos junto con ese filtrado, es posible rellenarlos de varias maneras. El m칠todo <code>fillna()</code> es el m치s utilizado en <code>Pandas</code>. Utilizar <code>fillna()</code> con una constante reemplaza los valores faltantes con ese valor establecido como constante.

```python
recaudacion = pd.read_html(URL, encoding='ISO-8859-1')  ## creacion de dataFrama de URL
df = recaudacion[0]
df.head()
```

```python
df.fillna(0).head() ## llenar columnas faltantes
```

<p align="justify"> 游 Usando <code>fillna()</code> con un diccionario, se puede establecer un valor de relleno diferente para cada columna indicada en el diccionario.

```python
df.fillna({231:0, 232:0.5, 233:1, 234:1.5}).head() ## rellena con diccionario de columna y valor
```

<p align="justify"> 游 Algunos par치metros
<ul>
<li><code>value()</code> valor u objeto similar a un diccionario para completar los valores que faltan</li>
<li><code>method()</code> m칠todo de interpolaci칩n <code>bfill</code> rellena hacia atr치s, <code>ffill</code> rellena hacia adelante, y por defecto el valor establecido es <code>None</code></li>
<li><code>axis()</code> eje para rellenar, siendo filas o columas el valor indicado</li>
<li><code>limit()</code> para el llenado hacia adelante y hacia atr치s, es el  n칰mero m치ximo de per칤odos consecutivos para llenar</li>
</ul>

 ## **<font color="#07a8ed">Transformar datos usando una funci칩n o mapeo</font>**

<p align="justify"> Para muchos conjuntos de datos, es posible realizar alguna transformaci칩n basada en los valores de una matriz, serie o columna en un <code>DataFrame</code>.

```python
df
```

```python
df.dropna(axis=1, how="any", inplace=True)  ###
df.head()
```

<p align="justify"> 游 Supongamos que deseamos agregar una columna que indique con n칰meros el mes resultante de la columna Meses. Previo a ello, vamos a poner nombres de columnas y eliminar el indice cero.

```python
df.columns = df.iloc[1]  # ubicacion [1]
df.columns.name = None  ## nombres de columnas
df.drop([0,1], axis=0, inplace=True) ## borro fila, en fila
df.reset_index(drop=True, inplace=True)  ## reseteo
df = df.query("Meses != 'Totales'")   ## filtro en columna meses reseteo
df.head()
```

```python
df.head()
```

```python
df.tail()
```

<p align="justify"> 游 El m칠todo <code>map()</code> en una serie acepta una funci칩n o un objeto similar a un diccionario que contiene un mapeo para hacer la transformaci칩n de valores que estamos requiriendo.



```python
df.head()
```

```python
df.tail()
```

```python
Mes = {"Enero":1, "Febrero":2, "Marzo":3, "Abril":4, "Mayo":5, "Junio":6,
       "Julio":7, "Agosto":8, "Septiembre":9, "Octubre":10, "Noviembre":11, "Diciembre":12}
```

```python
df.Meses = df.Meses.map(Mes)  ### reemplazo columna meses por el mapeo de meses con el diccionario
df.head()
```

<p align="justify"> 游 Otra forma:




```python
df.head()
```

```python
Meses = {1:"Enero", 2:"Febrero", 3:"Marzo", 4:"Abril", 5:"Mayo", 6:"Junio",
         7:"Julio", 8:"Agosto", 9:"Septiembre", 10:"Octubre", 11:"Noviembre", 12:"Diciembre"}
```

```python
def meses(x):  ### funcion de meses
    return Meses[x]
```

```python
df.Meses = df.Meses.map(meses)
df
```

<p align="justify"> 游 Otra forma:


```python
df.head()
```

```python
Mes = {"Enero":1, "Febrero":2, "Marzo":3, "Abril":4, "Mayo":5, "Junio":6,
       "Julio":7, "Agosto":8, "Septiembre":9, "Octubre":10, "Noviembre":11, "Diciembre":12}
```

```python
df.Meses = df.Meses.map(lambda x: Mes[x])
df
```

 ## **<font color="#07a8ed">Sustituci칩n de datos</font>**

<p align="justify"> Completar los datos que faltan con el m칠todo <code>fillna()</code> es un caso especial de reemplazo general. Como hemos visto, <code>map()</code> se puede usar para modificar un subconjunto de valores en un objeto, pero <code>replace()</code> proporciona una forma m치s simple y flexible de hacerlo.

```python
df.head()
```

```python
Meses = {1:"Enero", 2:"Febrero", 3:"Marzo", 4:"Abril", 5:"Mayo", 6:"Junio",
         7:"Julio", 8:"Agosto", 9:"Septiembre", 10:"Octubre", 11:"Noviembre", 12:"Diciembre"}
```

 ## **<font color="#07a8ed">Discretizaci칩n</font>**

<p align="justify"> Los datos continuos a menudo se discretizan o se separan en "contenedores" para su an치lisis. Supongamos que tenemos datos sobre un grupo de personas en un estudio comercial y deseamos agruparlos en grupos de edad discretos




```python
edad = [20, 22, 25, 27, 21, 23, 37, 31, 61, 45, 41, 32]
```

<p align="justify"> 游 Divid치moslos en grupos de 18 a 25, de 26 a 35, de 36 a 60 y, finalmente, de 61 a침os o m치s. Para hacerlo, tenemos que usar <code>cut()</code>


```python
contenedores = [18, 25, 35, 60, 100]
```

```python
categorias = pd.cut(edad, contenedores) ## pandas metodo cut contenedores
```

```python
categorias
```

<p align="justify"> 游 Cada contenedor se identifica mediante un tipo de valor de intervalo especial que contiene el l칤mite inferior y superior de cada contenedor


```python
pd.value_counts(categorias)
```

```python
categorias.value_counts()
```
